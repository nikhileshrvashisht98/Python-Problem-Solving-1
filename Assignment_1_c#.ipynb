```
//Question 3
// Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
// sides of a triangle and returns the area of the triangle. By Heron's formula, the area
// of a triangle with side lengths a, b, and c is (square root of (s(s − a)(s −b)(s −c))), where
// s = (a + b + c)/2.
// >>> triangleArea(2,2,2)
// 1.7320508075688772

// ###code starts here###
using System;

public class Triangle
{
    double triangleArea(float a,float b,float c)
    {
        double s = (a+b+c)/2;
        double area  = Math.Sqrt(s * (s - a) *(s - b) * (s - c));
        return area;
    }
    public static void Main(string[] args)
    {
        Triangle triangle = new Triangle();
        Console.Write("Enter the length of 1st side:");
        double side1 = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter the length of 2nd side:");
        double side2 = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter the length of 3rd side:");
        double side3 = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine(triangle.triangleArea(5,6,7));
    }
}
```

```
// Question 4 C#
// Write a program in C# Sharp to separate odd and even integers in separate arrays.
// Go to the editor
// Test Data :
// Input the number of elements to be stored in the array :5
// Input 5 elements in the array :
// element - 0 : 25
// element - 1 : 47
// element - 2 : 42
// element - 3 : 56
// element - 4 : 32
// Expected Output:
// The Even elements are:
// 42 56 32
// The Odd elements are :
// B
// 25 47

// ###code starts here###
    
using System;

public class Odd_Even
{
    public static void Main(string[] args)
    {
        Console.Write("Input the number of elements to be stored in the array : ");
        int len = Convert.ToInt32(Console.ReadLine());
        Console.WriteLine($"Input {len} elements in the array :");
        int []arr = new int[len];
        int even_index = 0; int odd_index = 0;
        int []even_arr = new int[len];
        int []odd_arr = new int[len];
        Console.WriteLine("\n");
        for (int i = 0;i<len;i++)
        {
            Console.Write($"element - {i} :");
            arr[i] = Convert.ToInt32(Console.ReadLine());
            if(arr[i]%2==0)
            {
                even_arr[even_index] = arr[i]; even_index++;
            }
            else
            {
                odd_arr[odd_index] = arr[i]; odd_index++;
            }
        }
        Console.WriteLine($"The Even elements are: ");
        for (int i = 0;i<even_index;i++)
        {
            Console.Write($"{even_arr[i]} ");
        }
        Console.WriteLine($"\nThe Odd elements are: ");
        for (int i = 0;i<odd_index;i++)
        {
            Console.Write($"{odd_arr[i]} ");
        }
    }
}

```

```
// Question 5 C#
// a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False
// dependingonwhetherthepoint(x,y)liesintherectanglewithlowerleft
// corner (x1,y1) and upper right corner (x2,y2).
// >>> inside(1,1,0,0,2,3)
// True
// >>> inside(-1,-1,0,0,2,3)
// False
// b. Use function inside()from part a. to write an expression that tests whether
// the point (1,1) lies in both of the following rectangles: one with lower left
// corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower
// left corner (0.5, 0.2) and upper right corner (1.1, 2)


// ###code starts here###
using System;

public class DotNetExpressions
{
    static Boolean  inside(double x, double y,double x1, double y1, double x2, double y2)
    {
        if ((x>=x1) && (x<=x2) && (y>=y1) && (y<=y2))
        {
            return true; 
        }
        else
        {
            return false;
        }
    }
    public static void Main(string[] args)
    {
        DotNetExpressions a = new DotNetExpressions();
        Console.WriteLine("Enter x,y co-ordinates of the lower left corner");
        double x1 = Convert.ToDouble(Console.ReadLine());
        double y1 = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Enter x,y co-ordinates of the top right corner");
        double x2 = Convert.ToDouble(Console.ReadLine());
        double y2 = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Enter x,y co-ordinates to be checked");
        double x = Convert.ToDouble(Console.ReadLine());
        double y = Convert.ToDouble(Console.ReadLine());
        Boolean abc = DotNetExpressions.inside(x,y,x1,y1,x2,y2);
        Console.WriteLine($"Whether the co-ordinates provided lies in the rectangle: "+abc);
        Boolean part_a_1 = DotNetExpressions.inside(1,1,0,0,2,3);
        Console.WriteLine($"Whether the co-ordinates inside(1,1,0,0,2,3) provided lies in the rectangle: "+part_a_1);
        Boolean part_a_2 = DotNetExpressions.inside(-1,-1,0,0,2,3);
        Console.WriteLine($"Whether the co-ordinates inside(-1,-1,0,0,2,3) provided lies in the rectangle: "+part_a_2);
        Boolean condition_1 = DotNetExpressions.inside(1,1,0.3,0.5,1.1,0.7);
        Boolean condition_2 = DotNetExpressions.inside(1,1,0.5,0.2,1.1,2);
        Console.WriteLine("Answer for the B part");
        if ((condition_1) && (condition_2))
        {
            Console.WriteLine("Point lies in both the rectangle");
        }
        else if ((condition_1)&&(!condition_2))
        {
            Console.WriteLine("Point  lies in 1st rectangle with the co-ordinates lower left corner (0.3, 0.5) and upper right corner (1.1, 0.7)\n but not in second rectangle");
        }
        else if ((!condition_1)&&(condition_2))
        {
            Console.WriteLine("Point  lies in 2nd rectangle with co-ordinates left corner (0.5, 0.2) and upper right corner (1.1, 2)\n but not in 1st rectangle");
        }
        else
        {
            Console.WriteLine("Point does not lie in both the rectangle");
        }
    }
}


```
